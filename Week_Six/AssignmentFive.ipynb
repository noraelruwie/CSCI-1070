#  (6 pts) Perform pre-processing on the dataset. Clearly use markdown to indicate each pre-processing step you are using. Note: you should use at least three different techniques. Depending on your dataset you may need to create new columns from the existent data you have.

```python
import pandas as pd 
import numpy as np

from sklearn.preprocessing import MinMaxScaler

titanic_df = pd.read_csv('titanic.csv')

titanic_df




titanic_df['Sex'] = titanic_df['Sex'].map({'female': 0, 'male': 1})


titanic_df['Age'] = titanic_df['Age'].replace('median_age', np.nan)

titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())

titanic_df['Embarked'] = titanic_df['Embarked'].fillna('Unknown')

titanic_df['Family_Size'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1

titanic_df = titanic_df.drop(['SibSp','Parch'], axis =1) 

titanic_df
```

# (3 pts) If using the Lending Club dataset create subplot histograms to show if loan_amounts differ by region. If using the Titanic dataset, create subplot histograms to show how fares differ by age_bin. 
# Summarize your findings in under 200 words.

```python
import matplotlib.pyplot as plt

bins = [0, 20, 40, 60, 80]
labels = ['0-20', '21-40', '41-60', '61+']

titanic_df['age_bin'] = pd.cut(titanic_df['Age'], bins=bins, labels=labels, right=False)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, ax in enumerate(axes.flatten()):
    age_group = labels[i]
    ax.hist(titanic_df[titanic_df['age_bin'] == age_group]['Fare'], bins=20, color='pink', edgecolor='black')
    ax.set_title(f'Fares for Age Group: {age_group}')
    ax.set_xlabel('Fare')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The histograms above represent the distribution of fares sold amongst each age group. You can see that the fares sell for different prices amongst different age groups, and that the cheaper fares are more people among every age group. The most expensive fares are most likely going to grant the guest a nicer room on board. It may look like the expensive fares are most popular in the age group 61+, but that age group has a lower frequency of fares sold. It would actually be more popular in the age group 41-60, and the age group 21-40. It looks like age group 0-20, and 21-40 prefer cheaper fares, although the latter does indulge in some expensive fares. It also looks like the titanic had the highest attendance in the age group 21-40, based on the amount of fares sold in that age group.
